```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

## Family Photo

```python
father = cv2.imread("tomcrussie.png")
mother = cv2.imread("katieholme.png")
child = cv2.imread("suricrusie.png")
```

```python
fig = plt.figure(figsize=(15, 15))

ax1 = fig.add_subplot(1,3,1)
plt.imshow(father[:,:,::-1]); plt.axis("off")

ax1 = fig.add_subplot(1,3,2)
plt.imshow(child[:,:,::-1]); plt.axis("off")

ax1 = fig.add_subplot(1,3,3)
plt.imshow(mother[:,:,::-1]); plt.axis("off")

plt.show()
```

## Parental Look-Alike Facial Recognition

```python
paternal_result = DeepFace.verify(img1_path="tomcrussie.png", 
                                  img2_path="suricrusie.png", 
                                  model_name="Facenet", 
                                  detector_backend="mtcnn",
                                  distance_metric="euclidean"
                                 )

maternal_result = DeepFace.verify(img1_path="katieholme.png", 
                                  img2_path="suricrusie.png", 
                                  model_name="Facenet", 
                                  detector_backend="mtcnn",
                                  distance_metric="euclidean"
                                 )
```

```python
fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_subplot(1,2,1)
plt.imshow(father[:,:,::-1]); plt.axis("off")

ax1 = fig.add_subplot(1,2,2)
plt.imshow(child[:,:,::-1]); plt.axis("off")

plt.show()

print(paternal_result)
```

```python
fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_subplot(1,2,1)
plt.imshow(mother[:,:,::-1]); plt.axis("off")

ax1 = fig.add_subplot(1,2,2)
plt.imshow(child[:,:,::-1]); plt.axis("off")

plt.show()

print(maternal_result)
```

```python
if paternal_result["distance"] < maternal_result["distance"]:
    print("the suricrussie is looking more like to her father tomcrussie")
else:
    print("the suricrusie is looking more like to her mother katieholmes")
```
